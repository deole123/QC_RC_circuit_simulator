```python
%matplotlib inline
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
import random
import string

sim = AerSimulator()

password = ""
#Genrate random characters 5 times
for i in range(5):
    qc = QuantumCircuit(3)
    qc.h(0)
    qc.h(1)
    qc.h(2)
    qc.measure_all()

    compiled = transpile(qc, sim)
    result = sim.run(compiled, shots = 1).result()
    counts = result.get_counts()
    binary_number = list(counts.keys())[0]
    characters = string.ascii_uppercase + string.digits
    password += random.choice(characters)

#Output
print("Quantum Random Passwod: ", password)
```
