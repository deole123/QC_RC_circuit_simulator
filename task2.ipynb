```python
import numpy as np
from qiskit_algorithms.linear_solvers.hhl import HHL
from qiskit.quantum_info import Statevector
import scipy.linalg

# 1. Define the integer system A and B
# A is a symmetric 2x2 integer matrix
A = np.array([[2, 1], 
              [1, 2]])

# B is our target vector
b = np.array([1, 0])

# 2. Initialize the HHL solver
hhl_solver = HHL()

# 3. Build the circuit and solve the system
solution = hhl_solver.solve(A, b)

# 4. Verification and Output
print("Quantum circuit successfully built!")
print(f"Quantum calculated Euclidean norm: {solution.euclidean_norm:.4f}")

# For a reality check, let's solve it classically using SciPy to compare
classical_solution = scipy.linalg.solve(A, b)
classical_norm = np.linalg.norm(classical_solution)
print(f"Classical exact Euclidean norm: {classical_norm:.4f}")

# 5. Viewing the Circuit (Optional)
# If you want to see the actual gates Qiskit generated for you:
# print(solution.state.draw())
```
