```python
%matplotlib inline
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector

#Even random number b/w 0 and 50
qc = QuantumCircuit(6)
qc.h(0)
qc.h(1)
qc.h(2)
qc.h(3)
qc.h(4)
qc.h(5)

#Block Vector visualization
#state = Statevector.from_instruction(qc)
#plot_bloch_multivector(state)

qc.measure_all()

#Simulate
sim = AerSimulator()
#Corrected line: Use the global transpile function
compiled = transpile(qc, sim)
result = sim.run(compiled, shots = 1).result()
counts = result.get_counts()
binary_number = list(counts.keys())[0]
even_number = (int(binary_number, 2) % 51)
#Output
qc.draw(output = 'mpl')
print("Measurement cicruit: ", even_number)
o=plot_histogram(counts)
plt.show()
```
